```python
data.loc[:,'大区']
```

```python
def change_value(x):
    if x['大区']==1:
        return '一大区'
    if x['大区']==2:
        return '二大区'
    if x['大区']==3:
        return '三大区'
    if x['大区']==4:
        return '四大区'
```

```python
data.loc[:,'大区'] = data.apply(change_value,axis=1)
```
