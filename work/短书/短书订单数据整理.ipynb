# 获取订单数据

# 获取推广订单数据

```python
tuiOrderPath = './datas/订单和推广员/2020-04-13推广纪录.xlsx'
```

```python
tuiOrderData = pd.read_excel(tuiOrderPath)
```

```python
tuiOrderData
```

# 合并订单和推广订单数据

```python
orer_tuiguangOrder_data = pd.merge(orderData,tuiOrderData,left_on='订单编号',right_on='订单号',how='left')
```

```python
orer_tuiguangOrder_data
```

```python
complete_order_data = orer_tuiguangOrder_data.loc[orer_tuiguangOrder_data['订单状态']=='已完成',:].copy()
```

```python
complete_order_data
```

```python
# 删除不必要的列
complete_order_data.drop(['支付订单号','支付流水号','卖家','支付类型','冻结','交易时间_x','卖家备注','团编号'],axis=1,inplace=True)
```

```python
complete_order_data
```

```python
# 数据导出
# complete_order_data.to_excel('./datas/order_clean.xlsx',index=False)
```

```python
order_name_data = pd.merge(complete_order_data,nameData,left_on='推广员昵称',right_on='微信名称',how='left')
```

```python
order_name_data
```

```python
# order_name_data.to_excel('./datas/order_name_clean.xlsx',index=False)
```
