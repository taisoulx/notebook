```python
import pandas as pd
```

```python
fPath = './datas/tuiguangyuan.xlsx'
```

```python
tuiData = pd.read_excel(fPath,error_bad_lines=False)
```

```python
tuiData.head()
```

```python
tuiData.count
```

```python
namePath = './datas/推广员对应合伙人真实姓名与区域 1.xlsx'
```

```python
nameData = pd.read_excel(namePath)
```

```python
nameData
```

```python
data_tuiguang = pd.merge(tuiData,nameData,left_on='微信名称',right_on='推广员昵称',how='left')
```

```python
data_tuiguang
```

```python
data_tuiguang.drop(['加入时间','来源/加入方式','Unnamed: 0'],axis=1,inplace=True)
```

```python
data_tuiguang
```

```python
duplicate_row = data_tuiguang.duplicated(subset=['微信名称'],keep='first') 
```

```python
duplicate_row
```

```python
def get_bool(x):
    if x==True:
        return False
    return True
```

```python
juge = duplicate_row.apply(get_bool)
```

```python
data=data_tuiguang.loc[ juge,:]
```
